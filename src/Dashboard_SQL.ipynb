
#  1️⃣ LAST PRICE — Latest collected value
# 
#  ❓ Question: What is the latest recorded Bitcoin price?


```python
%sql
 SELECT 
  valor_brl  AS last_price,
  timestamp  AS last_timestamp

 FROM pipeline_api_bitcoin.biticoin_delta.btc_data
 WHERE criptomoeda =  'BTC'  
 ORDER BY timestamp DESC
 LIMIT 1
```

#  2️⃣ MAX PRICE — Highest value ever recorded + when it occurred
# ❓ Question: What was the highest price ever recorded and when did it happen?


```python
%sql
 SELECT 
  valor_brl  AS max_price,
  timestamp  AS max_timestamp

 FROM pipeline_api_bitcoin.biticoin_delta.btc_data
 WHERE criptomoeda =  'BTC'  
 ORDER BY valor_brl DESC,timestamp DESC
 LIMIT 1
```

# 3️⃣ MIN PRICE — Lowest value ever recorded + when it occurred
# ❓ Question: What was the lowest price ever recorded and when did it happen?


```python
%sql
 SELECT 
  valor_brl  AS min_price,
  timestamp  AS min_timestamp

 FROM pipeline_api_bitcoin.biticoin_delta.btc_data
 WHERE criptomoeda =  'BTC'  
 ORDER BY valor_brl ASC,timestamp DESC
 LIMIT 1
```
